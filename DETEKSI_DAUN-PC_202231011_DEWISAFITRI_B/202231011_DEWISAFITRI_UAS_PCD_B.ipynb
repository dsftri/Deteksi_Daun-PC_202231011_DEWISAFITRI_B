```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = 'jambu.jpg'
image = cv2.imread(image_path)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)


lower_brown = np.array([0, 70, 70])
upper_brown = np.array([20, 255, 255])

mask_fruit = cv2.inRange(hsv, lower_brown, upper_brown)

segmented_fruit = cv2.bitwise_and(image_rgb, image_rgb, mask=mask_fruit)

lower_green = np.array([35, 20, 50])
upper_green = np.array([85, 255, 255])

mask_leaf = cv2.inRange(hsv, lower_green, upper_green)
segmented_leaf = cv2.bitwise_and(image_rgb, image_rgb, mask=mask_leaf)



plt.figure(figsize=(20, 20))

plt.subplot(2, 2, 1)
plt.imshow(image_rgb)
plt.title("Gambar asli")
plt.axis('on')

plt.subplot(2, 2, 2)
plt.imshow(mask_fruit, cmap='gray')
plt.title("Mask jambu")
plt.axis('on')

plt.subplot(2, 2, 3)
plt.imshow(segmented_fruit)
plt.title("Segmentasi jambu")
plt.axis('on')

plt.subplot(2, 2, 4)
plt.imshow(segmented_leaf)
plt.title("Segmentasi daun")
plt.axis('on')

plt.show()
```
